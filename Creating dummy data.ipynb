### [Problem 1] Creating random numbers

```python
import numpy as np
mean = [-3, 0]
covar = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
rand_num = np.random.multivariate_normal(mean, covar, 500)
print(rand_num)
rand_num.shape
```

### [Problem 4] Addition of data

```python
import numpy as np
mean1 = [0, -3]
covar = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
rand_num1 = np.random.multivariate_normal(mean1, covar, 500)
print(rand_num1)
rand_num1.shape
```

```python
plt.title("Two in one scatter plot")
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(rand_num[:,0],rand_num[:, 1], s = 20, label = '0', marker = 'o')

plt.scatter(rand_num1[:,0], rand_num1[:, 1], s = 20, label = '1', marker = 'o')
plt.legend()
plt.show
```

### [Problem 5] Data combination

```python
# Using concatenate
data_combine = np.concatenate((rand_num, rand_num1), axis = 0)
print(data_combine)
data_combine.shape
```

```python
# using vstack
combined_data = np.vstack((rand_num, rand_num1))
combined_data.shape
```
